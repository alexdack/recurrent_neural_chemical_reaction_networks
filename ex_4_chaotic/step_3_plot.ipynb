```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import matplotlib as mpl
from cycler import cycler
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import matplotlib.transforms as mtransforms
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.lines as mlines
```

```python
CheRR_Simulation=[]
with open('data/hasting_powell_RNCRN_traj.csv',newline='') as csvfile:
    data = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in data:
        CheRR_Simulation.append([float(datapoint) for datapoint in row[0].split(',')])
        
CheRR_Simulation=np.asarray(CheRR_Simulation) 


NonKin_Simulation=[]
with open('data/hasting_powell_non_kinetic_traj.csv',newline='') as csvfile:
    data = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in data:
        NonKin_Simulation.append([float(datapoint) for datapoint in row[0].split(',')])
        
NonKin_Simulation=np.asarray(NonKin_Simulation) 
```

```python
mpl.rcParams['lines.linewidth'] = 0.5
mpl.rcParams['font.size'] = 30
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.serif'] = 'Times New Roman'
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'serif'
plt.rcParams['axes.prop_cycle'] = cycler(alpha=[1])
mpl.rcParams['figure.figsize'] = [24 , 14]
mpl.rcParams['text.usetex'] = False
mpl.rcParams['legend.frameon'] = False

ColList = ['#1F13FF','#FF00D0', '#2A2A72', '#bbbdbf']


tickfontsize=30
labelfontsize=35
titlefontsize=35 

fig, axs = plt.subplot_mosaic([['(a)', '(a)', '(a)', '(b)', '(b)', '(b)'],
                               ['(c)', '(c)', '(d)', '(d)', '(e)', '(e)']], gridspec_kw={ 'height_ratios':[5,1]}, layout='constrained')

fig.tight_layout()
for label, ax in axs.items():
    # label physical distance to the left and up:
    trans = mtransforms.ScaledTranslation(-40/72, 14/72, fig.dpi_scale_trans)
    ax.text(0.0, 1.0, label, transform=ax.transAxes + trans,
            fontsize='medium', va='bottom', fontfamily='serif', ha='right')

ss = axs['(a)'].get_subplotspec()
axs['(a)'].remove()
axs['(a)'] = fig.add_subplot(ss, projection='3d')
axs['(a)'].text2D(0.0, 1.0,'(a)', transform=axs['(a)'].transAxes + trans,
            fontsize='medium', va='bottom', fontfamily='serif', ha='right')

ss = axs['(b)'].get_subplotspec()
axs['(b)'].remove()
axs['(b)'] = fig.add_subplot(ss, projection='3d')
axs['(b)'].text2D(0.0, 1.0,'(b)', transform=axs['(b)'].transAxes + trans,
            fontsize='medium', va='bottom', fontfamily='serif', ha='right')

ax1 = axs['(a)']
ax1.zaxis.set_rotate_label(False)  
ax1.plot3D(NonKin_Simulation[:,1], NonKin_Simulation[:,2], NonKin_Simulation[:,3], '-' ,color=ColList[0])
ax1.set_xticks([0,1])
ax1.set_yticks([0,1])
ax1.set_zticks([0,1/2])
ax1.set_xticklabels(['0','1'],fontsize=tickfontsize)
ax1.set_yticklabels(['0','1'],fontsize=tickfontsize)
ax1.set_zticklabels(['0','1/2'],fontsize=tickfontsize)
ax1.tick_params(axis='both', which='major')
ax1.set_xlabel(r'$\bar{x}_1$', fontsize=labelfontsize )
ax1.set_ylabel(r'$\bar{x}_2$', fontsize=labelfontsize )
ax1.set_zlabel(r'$\bar{x}_3$', rotation=0, fontsize=labelfontsize )
ax1.set_title(r'Target system', loc='center', fontsize=labelfontsize)
ax1.view_init(20, -120)
ax1.set_facecolor('white') 
ax1.grid(False) 
ax1.w_xaxis.pane.fill = False
ax1.w_yaxis.pane.fill = False
ax1.w_zaxis.pane.fill = False
ax1.tick_params(bottom=True, top=True, left=True, right=True)
ax1.tick_params(bottom=True, top=True, left=True, right=True, which='minor')
ax1.xaxis.set_minor_locator(AutoMinorLocator(10))
ax1.yaxis.set_minor_locator(AutoMinorLocator(10))
ax1.zaxis.set_minor_locator(AutoMinorLocator(5))
ax1.tick_params(axis="x", direction="in")
ax1.tick_params(axis="x", direction="in", which="minor")
ax1.tick_params(axis="y", direction="in")
ax1.tick_params(axis="y", direction="in", which="minor")

ax2 = axs['(b)']
ax2.zaxis.set_rotate_label(False)  # disable automatic rotation
ax2.plot3D(CheRR_Simulation[:,1], CheRR_Simulation[:,2], CheRR_Simulation[:,3], '-' ,color=ColList[1])
ax2.set_xticks([0,1])
ax2.set_yticks([0,1])
ax2.set_zticks([0,0.5])
ax2.set_xticklabels(['0','1'],fontsize=tickfontsize)
ax2.set_yticklabels(['0','1'],fontsize=tickfontsize)
ax2.set_zticklabels(['0','1/2'],fontsize=tickfontsize)
ax2.tick_params(axis='both', which='major' )
ax2.set_xlabel(r'$x_1$' , fontsize=labelfontsize)
ax2.set_ylabel(r'$x_2$', fontsize=labelfontsize )
ax2.set_zlabel(r'$x_3$', rotation=0, fontsize=labelfontsize)
ax2.set_title(r'RNCRN approximation', loc='center', fontsize=titlefontsize)
ax2.set_xlim3d(0,1)
ax2.set_ylim3d(0,1)
ax2.set_zlim3d(0,0.5)
ax2.view_init(20, -120)
ax2.set_facecolor('white') 
ax2.grid(False) 
ax2.w_xaxis.pane.fill = False
ax2.w_yaxis.pane.fill = False
ax2.w_zaxis.pane.fill = False
ax2.tick_params(bottom=True, top=True, left=True, right=True)
ax2.tick_params(bottom=True, top=True, left=True, right=True, which='minor')
ax2.xaxis.set_minor_locator(AutoMinorLocator(10))
ax2.yaxis.set_minor_locator(AutoMinorLocator(10))
ax2.zaxis.set_minor_locator(AutoMinorLocator(5))


ax3 = axs['(c)']
ax3.plot(NonKin_Simulation[:,0], NonKin_Simulation[:,1],'-' ,color=ColList[0], linewidth=2)
ax3.plot(CheRR_Simulation[:,0], CheRR_Simulation[:,1],'--' ,color=ColList[1], linewidth=2)
ax3.set_xticks(np.arange(0,120,20))
ax3.set_yticks([0,0.5,1,1.5])
ax3.tick_params(axis='both', which='major' )
ax3.set_xlabel(r'Time $t$', fontsize=labelfontsize )
ax3.set_ylabel(r'$x_1(t)$', fontsize=labelfontsize )
ax3.set_xlim(0,90)
ax3.set_ylim(0, 1.1)
ax3.set_title(r'$x_1(t)$ trajectories', loc='center', fontsize=titlefontsize,  pad=15.5)
ax3.tick_params(bottom=False, top=True, left=True, right=True)
ax3.tick_params(bottom=True, top=True, left=True, right=True, which='minor')
ax3.xaxis.set_minor_locator(AutoMinorLocator(10))
ax3.yaxis.set_minor_locator(AutoMinorLocator(5))
ax3.tick_params(axis="x", direction="in", labelsize=tickfontsize)
ax3.tick_params(axis="x", direction="in", which="minor")
ax3.tick_params(axis="y", direction="in", labelsize=tickfontsize)
ax3.tick_params(axis="y", direction="in", which="minor")
ax3.fill([0, 0, 300, 300], [1, 2, 2, 1], color = 'gray', alpha=0.1)
ax3.fill([0, 0, 300, 300], [1, 2, 2, 1], color = 'None', alpha=0.1, edgecolor=ColList[3], hatch="/")
ax3.fill([0, 0, 300, 300], [-0.3, 0.01, 0.01, -0.3], color = 'gray', alpha=0.1)
ax3.fill([0, 0, 300, 300], [-0.3, 0.01, 0.01, -0.3], color = 'None', alpha=0.1, edgecolor=ColList[3], hatch="/")
ax3.hlines(0.01, -2, 300, facecolor ='white', edgecolor='k', linewidth=1.5, linestyle='dashdot' )
ax3.hlines(1, -2, 300, facecolor ='white', edgecolor='k', linewidth=1.5, linestyle='dashdot' )

ax4 = axs['(d)']
ax4.plot(NonKin_Simulation[:,0], NonKin_Simulation[:,2],'-' ,color=ColList[0], linewidth=2)
ax4.plot(CheRR_Simulation[:,0], CheRR_Simulation[:,2],'--' ,color=ColList[1], linewidth=2)
ax4.set_xticks(np.arange(0,120,20))
ax4.set_yticks([0,0.5,1,1.5])
ax4.tick_params(axis='both', which='major' )
ax4.set_xlabel(r'Time $t$', fontsize=labelfontsize )
ax4.set_ylabel(r'$x_2(t)$', fontsize=labelfontsize )
ax4.set_xlim(0,90)
ax4.set_ylim(0, 1.1)
ax4.set_title(r'$x_2(t)$ trajectories', loc='center', fontsize=titlefontsize,  pad=15.5)
ax4.tick_params(bottom=False, top=True, left=True, right=True)
ax4.tick_params(bottom=True, top=True, left=True, right=True, which='minor')
ax4.xaxis.set_minor_locator(AutoMinorLocator(10))
ax4.yaxis.set_minor_locator(AutoMinorLocator(5))
ax4.tick_params(axis="x", direction="in", labelsize=tickfontsize)
ax4.tick_params(axis="x", direction="in", which="minor")
ax4.tick_params(axis="y", direction="in", labelsize=tickfontsize)
ax4.tick_params(axis="y", direction="in", which="minor")
ax4.fill([0, 0, 300, 300], [1, 2, 2, 1], color = 'gray', alpha=0.1)
ax4.fill([0, 0, 300, 300], [1, 2, 2, 1], color = 'None', alpha=0.1, edgecolor=ColList[3], hatch="/")
ax4.fill([0, 0, 300, 300], [-0.3, 0.01, 0.01, -0.3], color = 'gray', alpha=0.1)
ax4.fill([0, 0, 300, 300], [-0.3, 0.01, 0.01, -0.3], color = 'None', alpha=0.1, edgecolor=ColList[3], hatch="/")
ax4.hlines(0.01, -2, 300, facecolor ='white', edgecolor='k', linewidth=1.5, linestyle='dashdot' )
ax4.hlines(1, -2, 300, facecolor ='white', edgecolor='k', linewidth=1.5, linestyle='dashdot' )



ax5 = axs['(e)']
ax5.plot(NonKin_Simulation[:,0], NonKin_Simulation[:,3],'-' ,color=ColList[0], linewidth=2)
ax5.plot(CheRR_Simulation[:,0], CheRR_Simulation[:,3],'--' ,color=ColList[1], linewidth=2)
ax5.set_xticks(np.arange(0,120,20))
ax5.set_yticks([0,0.5,1,1.5])
ax5.tick_params(axis='both', which='major' )
ax5.set_xlabel(r'Time $t$', fontsize=labelfontsize )
ax5.set_ylabel(r'$x_3(t)$',fontsize=labelfontsize  )
ax5.set_xlim(0,90)
ax5.set_ylim(0, 1.1)
ax5.tick_params(bottom=True, top=True, left=True, right=True)
ax5.tick_params(bottom=True, top=True, left=True, right=True, which='minor')
ax5.xaxis.set_minor_locator(AutoMinorLocator(10))
ax5.yaxis.set_minor_locator(AutoMinorLocator(5))
ax5.tick_params(axis="x", direction="in", labelsize=tickfontsize)
ax5.tick_params(axis="x", direction="in", which="minor")
ax5.tick_params(axis="y", direction="in", labelsize=tickfontsize)
ax5.tick_params(axis="y", direction="in", which="minor")
ax5.fill([0, 0, 300, 300], [1, 2, 2, 1], color = 'gray', alpha=0.1)
ax5.fill([0, 0, 300, 300], [1, 2, 2, 1], color = 'None', alpha=0.1, edgecolor=ColList[3], hatch="/")
ax5.fill([0, 0, 300, 300], [-0.3, 0.01, 0.01, -0.3], color = 'gray', alpha=0.1)
ax5.fill([0, 0, 300, 300], [-0.3, 0.01, 0.01, -0.3], color = 'None', alpha=0.1, edgecolor=ColList[3], hatch="/")
ax5.hlines(0.01, -2, 300, facecolor ='white', edgecolor='k', linewidth=1.5, linestyle='dashdot' )
ax5.hlines(1, -2, 300, facecolor ='white', edgecolor='k', linewidth=1.5, linestyle='dashdot' )
ax5.set_title(r'$x_3(t)$ trajectories', loc='center', fontsize=titlefontsize,  pad=15.5)

plt.show()
fig.savefig('figures/figure_4.pdf', bbox_inches='tight')
```
